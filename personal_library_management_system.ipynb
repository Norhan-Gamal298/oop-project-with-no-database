```python
from abc import ABC, abstractmethod
class Book(ABC):
    def __init__(self,new_title,new_auther,new_genre,publication_year):
        self.__title=new_title
        self.__auther=new_auther 
        self.__genre=new_genre
        self.__year=publication_year
    def set_book(self,new_title,new_auther,new_genre,publication_year):
        try:
            self.__title=str(new_title)
        except Exception:
            print("datatype of what you entered is not correct,please enter correct data type")

        try:
            self.__auther=str(new_auther)
        except Exception:
            print("datatype of what you entered is not correct,please enter correct data type")
        try:
            self.__genre=str(new_genre)
        except Exception:
            print("datatype of what you entered is not correct,please enter correct data type")
        try:
            self.__year=int(publication_year)
        except Exception:
            print("datatype of what you entered is not correct,please enter correct data type")
    @abstractmethod
    def add_newbook(self):
        pass
    
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

def center_window(win, width=400, height=300):
    win.update_idletasks()
    x = (win.winfo_screenwidth() // 2) - (width // 2)
    y = (win.winfo_screenheight() // 2) - (height // 2)
    win.geometry(f"{width}x{height}+{x}+{y}")

class Library(Book):
    def __init__(self, new_title, new_auther, new_genre, publication_year):
        super().__init__(new_title, new_auther, new_genre, publication_year)
        self.book_collection={}    #key=title,value= list contains rest of data(auther,year,genre)
        self.value_list=[new_auther,new_genre, publication_year]
        self.key_list= new_title
        self.book_list=[]
    def add_newbook(self):
            self.book_collection[self.key_list]=self.value_list
            self._save()
            print( self.book_collection)
    def _save(self):
         with open("book_store.txt",'a') as f:
            for key,value in self.book_collection.items():
                f.write(f"\nTitle: {key} Author: {value[0]}, Genre: {value[1]}, Publication Year: {value[2]}")
         
    def remove_book(self,title):
         with open ("book_store.txt",'r') as f:
            lines=f.readlines()
            new_lines=[]
            for i in lines:
                if title not in i:
                    new_lines.append(i)
         with open ("book_store.txt",'w') as nf:
              nf.writelines(new_lines)
         if len(lines) > len(new_lines):
              print("the book is deleted successfully") 
         else:
            print("book not found")
    def look_for_book(self, title):
         with open ("book_store.txt",'r') as f:
            lines=f.readlines()
            new_lines=[]
            flag=False
            for i in lines:
                if title  in i:
                    new_lines.append(i)
                    flag=True
                    break
            if flag==True:
                return new_lines
            else:
                return "no book found!"  
    def view_books(self):
        try:
            with open ("book_store.txt",'r') as f:
                lines=f.readlines()    
                if not lines :
                    return "no books found!"
                return lines
        except FileNotFoundError:
            return "file not found!"
        


#tkinter GUI        
                   
def add_book():
    global library
    title = title_entry.get()
    auther = auther_entry.get()
    genre = genre_entry.get()
    year = year_entry.get()
    if title and auther and genre and year.isdigit():
        library = Library(title, auther, genre, int(year))
        library.add_newbook()
        messagebox.showinfo("Success", "Book added successfully!")
    else:
        messagebox.showerror("Error", "Please enter valid data.")

def remove_book():
    title = title_entry.get().strip()
    if not title:
        messagebox.showerror("Error", "Please enter a book title to remove.")
        return
    try:
        temp_lib = Library("", "", "", 0)
        temp_lib.remove_book(title)
        messagebox.showinfo("Success", f"Book '{title}' deleted (if it existed).")
    except FileNotFoundError:
        messagebox.showerror("Error", "No book store file found yet!")

def search_book():
    title = title_entry.get().strip()
    if not title:
        messagebox.showerror("Error", "Please enter a book title.")
        return
    try:
        temp_lib = Library("", "", "", 0)
        results = temp_lib.look_for_book(title)
        if isinstance(results, list) and results:
            messagebox.showinfo("Found", "\n".join(results))
        else:
            messagebox.showwarning("Not Found", f"No book found with title '{title}'")
    except FileNotFoundError:
        messagebox.showerror("Error", "No book store file found yet!")

def view_books():
    try:
        temp_lib = Library("", "", "", 0)
        results = temp_lib.view_books()
        if isinstance(results, list) and results:
            messagebox.showinfo("All Books", "\n".join(results))
        else:
            messagebox.showinfo("Info", results)
    except FileNotFoundError:
        messagebox.showerror("Error", "No book store file found yet!")

def clear_fields():
    title_entry.delete(0, tk.END)
    auther_entry.delete(0, tk.END)
    genre_entry.delete(0, tk.END)
    year_entry.delete(0, tk.END)

root = tk.Tk()
root.title("Library System")
root.configure(bg="#f0f4f8")
center_window(root, 800, 620)
root.resizable(False, False)

main_frame = ttk.Frame(root, padding="20 10 20 10")
main_frame.pack(fill=tk.BOTH, expand=True)

style = ttk.Style()
style.configure("TFrame", background="#34495E")
style.configure("TLabel", background="#34495E", foreground="black")
style.configure("TEntry", fieldbackground="black")
style.configure("TButton", background="#34495E", foreground="black")


main_frame.columnconfigure(0, weight=1)
main_frame.columnconfigure(1, weight=3)
for i in range(6):
    main_frame.rowconfigure(i, weight=1)
title_lbl = ttk.Label(main_frame, text="Personal Library Management system", font=("Arial", 22, "bold"))
title_lbl.grid(row=0, column=0, columnspan=2, pady=(0, 15))

ttk.Label(main_frame, text="Title:", style="TLabel", font=("Arial", 16)).grid(row=1, column=0, sticky="ew", pady=5, padx=(0,250))
title_entry = ttk.Entry(main_frame)
title_entry.grid(row=1, column=1, sticky="ew", pady=5, padx=(0,15))

ttk.Label(main_frame, text="Author:", style="TLabel", font=("Arial", 16)).grid(row=2, column=0, sticky="ew", pady=5, padx=(0,250))
auther_entry = ttk.Entry(main_frame)
auther_entry.grid(row=2, column=1, sticky="ew", pady=5, padx=(0,15))

ttk.Label(main_frame, text="Genre:", style="TLabel", font=("Arial", 16)).grid(row=3, column=0, sticky="ew", pady=5, padx=(0,250))
genre_entry = ttk.Entry(main_frame)
genre_entry.grid(row=3, column=1, sticky="ew", pady=5, padx=(0,15))

ttk.Label(main_frame, text="publication_year:", style="TLabel", font=("Arial", 16)).grid(row=4, column=0, sticky="ew", pady=5, padx=(0,250))
year_entry = ttk.Entry(main_frame)
year_entry.grid(row=4, column=1, sticky="ew", pady=5, padx=(0,15))

btn_frame = ttk.Frame(main_frame)
btn_frame.grid(row=5, column=0, columnspan=2, pady=15)

ttk.Button(btn_frame, text="Add Book", command=add_book).grid(row=0, column=0, padx=5)
ttk.Button(btn_frame, text="Remove Book", command=remove_book).grid(row=0, column=1, padx=5)
ttk.Button(btn_frame, text="Search Book", command=search_book).grid(row=0, column=2, padx=5)
ttk.Button(btn_frame, text="View All Books", command=view_books).grid(row=0, column=3, padx=5)
ttk.Button(btn_frame, text="Clear", command=clear_fields).grid(row=0, column=4, padx=5)

root.mainloop()
```
